```python
# MySQL 연동

from sqlalchemy import create_engine, text
import numpy as np
import pandas as pd

# 접속정보
# engine = create_engine('mysql+pymysql://[계정이름]:[계정비밀번호]@localhost:3306[접속 포트번호]/[데이터베이스 이름]?charset=utf8mb4')
engine = create_engine('mysql+pymysql://admin:admin1234@localhost:3306/example_db?charset=utf8mb4')

# 데이터 selection
# 책 제목에 여행이 들어가는 책
# USE example_db;

# SELECT bisbn, btitle, bauthor, bprice
# FROM books
# WHERE btitle LIKE '%여행%';
search_keyword = '여행'
sql = 'SELECT bisbn, btitle, bauthor, bprice FROM books WHERE btitle LIKE :keyword'

# With는 file처리나 network처리에서 open후 close를 자동으로 처리하기 위해서 사용
with engine.connect() as connection:
    result = connection.execute(text(sql), {
        'keyword' : f'%{search_keyword}%'
    })

    df = pd.DataFrame(result.fetchall(),
                      columns=result.keys())
    
display(df)

# MySQ을 통해 DataFrame을 생성했는데 이걸
# CSV로 저장해 볼게요!
df.to_csv('mybook.csv', index=False)

# CSV도 많이 이용하지만, JSON도 많이 이용
# JSON파일을 만들어서 DataFrame내용을 저장
with open('book_orient_record.json', 'w', encoding='utf-8') as file:
    df.to_json(file, force_ascii=False, orient='records') # 일반적으로 많이 사용

with open('book_orient_columns.json', 'w', encoding='utf-8') as file:
    df.to_json(file, force_ascii=False, orient='columns') # 일반적으로 많이 사용

# df.to_json('book_orient_index.json', force_ascii=False, orient='index')
```

```python
import numpy as np
import pandas as pd
import json
import urllib

key = 'c7e10851cd6000c340de454f617125bf' # 발급받은 인증키
targetDT = '20240901' # 조회할 날짜

movie_url = f'http://kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key={key}&targetDt={targetDT}'

load_page = urllib.request.urlopen(movie_url) # 접속 후 응답객체를 가져오는거예요!

json_page = json.loads(load_page.read()) # 응답객체의 내용을 읽어서 json형태로 변환

print(type(json_page)) # <class 'dict'>

print(json_page.keys())

# json_page['boxOfficeResult']['dailyBoxOfficeList'][0]['movieNm'] # '에이리언: 로물루스'
# DataFrame 만드는 방법
# data = {
#     '순위' : ['1', '2'],
#     '영화제목' : ['에일리언', '얼굴'],
#     '매출액' : [10000, 20000]
# }
movie_data = dict()
rank_list = []
title_list = []
sales_list = []

for tmp in json_page['boxOfficeResult']['dailyBoxOfficeList']:
    rank_list.append(tmp['rank'])
    title_list.append(tmp['movieNm'])
    sales_list.append(tmp['salesAmt'])

movie_data['순위'] = rank_list
movie_data['영화제목'] = title_list
movie_data['매출액'] = sales_list

df = pd.DataFrame(movie_data)
display(df)
```

```python
import numpy as np
import pandas as pd

# dictionary 직접 만들어서 DataFrame 생성
data = {
    '이름': ['홍길동', '신사임당', '강감찬', '아이유','이순신'],
    '학과': ['경영학과', '국어국문', '군사학과', '실용음악', '군사학과'],
    '학년': [1, 2, 2, 4, 3],
    '학점': [1.5, 2.0, 3.1, 1.1, 2.7]
}

df = pd.DataFrame(data,
                  columns=['학과', '이름', '학점', '학년', '등급'],
                  index=['one', 'two', 'three', 'four', 'five'])
# 숫자 index가 아닌 문자 index로 변경
# Nan값은 값이 존재하지 않는다는 의미

# 하나의 컬럼만 추출하지만 그 값을 변경할거예요
# year = df['학년'] # year는 Series

# year['one'] = 1000 # 컬럼 인덱싱한 Series의 값을 변경
# display(df)        # Series의 값을 변경하면 DataFrame의 값도 변경
#                    # 컬럼 인덱싱하면 View가 나와요!
#                    # 당연히 좋지 않아요! 문제를 야기시켜요!

# 명시적으로 View를 사용하지 않고 copy본을 이용하는 습관을 갖이자
year = df['학년'].copy() # View가 아닌 copy()를 이용해서 복사본을 만들어서 사용
year['one'] = 1000 # 복사본이기 때문에 원본 DataFrame에 영향 없음

df['학년'] = year # 변경된 복사본을 DataFrame에 다시 할당
display(df)        # Series의 값을 변경해도 DataFrame의 값은 변경되지 않음
```
