```python
import pandas as pd
import numpy as np

# dictionary 직접 만들어서 DataFrame 생성
data = {
    '이름': ['홍길동', '신사임당', '강감찬', '아이유','이순신'],
    '학과': ['경영학과', '국어국문', '군사학과', '실용음악', '군사학과'],
    '학년': [1, 2, 2, 4, 3],
    '학점': [1.5, 2.0, 3.1, 1.1, 2.7]
}

df = pd.DataFrame(data,
                  columns=['학과', '이름', '학점', '학년', '등급'],
                  index=['one', 'two', 'three', 'four', 'five'])

display(df)

# 학점이 1.5를 초과하는 학생의 이름과 학점을 DataFrame으로 가져오기
# df.loc[행, 열]
df0 = df.loc[df['학점'] > 1.5, ['이름', '학점']]
display(df0)

# 1. 이름이 강감찬인 사람을 찾아서 이름과 학점을 DataFrame으로 출력
one = df.loc[df['이름'] == '강감찬', ['이름', '학점']]
display(one)
# 2. 학점이 1.5초과 2.5 미만인 사람을 찾아서 학과 이름 학점을 출력
two = df.loc[(df['학점'] > 1.5) & (df['학점'] < 2.5), ['학과', '이름', '학점']]
display(two)

# 3. 학점이 3.0을 초과하는 사람을 찾아서 등급을 'A'로 설정
three = df.loc[df['학점'] > 3.0, '등급'] = 'A'
display(df)

```

```python
# 1. shape
print(df.shape) # (398, 9) 무조건 tuple 출력 행은 398, 열은 9개

# 2. head(), tail() => 데이터프레임의 상위 n개, 하위 n개를 추출하는 함수(기본값 5)
display(df.head()) # 상위 5개
display(df.tail()) # 하위 5개

# 3. info() => 데이터프레임의 기본 정보를 출력해줘요!
#              기본정보를 이용해서 결측치(Missing Value, NaN)나
#              데이터 타입 정보를 확인할 수 있어요
df.info()
```

```python
# 4. count()
# 데이터 프레임이 가지고 있는 각 열의 데이터 개수를 Series로 리턴
# NaN은 제외하고 개수를 세요! => Missing Value가 있는지 확인할 때 쓰면 좋아요!
print(df.count())
```

```python
# 5. value_count()
# DataFrame의 함수가 아니라 Series가 가지고 있는 함수
# 고유값의 개수 => 유니크한 값이 어떤게 있고 각각 몇개씩 값을 가지고 있는지.
df['origin'].value_counts()
```

```python
# 6. unique()
# DataFrame이 아니라 Series에 적용하는 함수
# 어떤 값이 존재하는지 확인
df['origin'].unique()
```

```python
# 7. isin() : 특정값이 포함되어 있는지를 boolean으로 리턴
#             일반적으로 mask를 만들때 많이 사용

# 우리 MPG데이터는 3개국에서(미국(1), EU(2), 일본(3))
# 정보를 가지고 있는 데이터
# 제조국이 미국과 일본인 자동차만 추출
df.loc[df['origin'].isin([1,3]), :]
```

```python
# 8. 가장 흔하게 사용하는 함수 중 하나가 sort()
#    사용할 때 살짝 조심해야 하는 부분

# random하게 데이터를 만들어 볼꺼예요(정수형 난수)
# 난수의 재현을 위해 seed를 잡고 시작
np.random.seed(42)

df = pd.DataFrame(np.random.randint(0, 10, (6,4)))
df.columns = ['A', 'B', 'C', 'D']
df.index = pd.date_range('20250101', periods=6)

# 이렇게 만들어진 DataFrame의 컬럼의 순서와 row index의 순서를
# 랜덤하게 바꿀꺼에요
# 원래 랜덤하게 섞는 함수는 numpy가 제공하는 shuffle이 있어요
# np.random.shuffle(df.index) # index가 가진 특성상 직접 수정이 안되요!
# 그래서 다른 방식을 사용 => 함수를 다른거 써야 해요!
random_date = np.random.permutation(df.index) # 랜덤하게 섞인 날짜를 얻어요!

# index와 colums을 재 설정할 수 있어요
# reindex()
new_df = df.reindex(index=random_date,
                    columns=['B', 'A', 'D', 'C'])
display(new_df)

# 컬럼을 오름차순으로 정렬
display(new_df.sort_index(axis=1, ascending=True))
# 행과 열을 정렬할 때는 따로따로 실행
# 위에 있는건 index를 이용한 정렬

# 값을 가지고 정렬 / 1차 정렬
display(new_df.sort_values(by='B'))

# 2차 정렬 방법
display(new_df.sort_values(by=['D', 'B']))
```

```python
import numpy as np
import pandas as pd

data = [[2, np.nan],
        [7, -3],
        [np.nan, np.nan],
        [1, -2]]
df = pd.DataFrame(data,
                  columns=['one', 'two'],
                  index=['a', 'b', 'c', 'd'])

# 가장 대표적인 집계함수 : sum()
# numpy의 sum()과 DataFrame의 sum() 약간 달라요
display(df)
print(df.sum()) # axis = 0 (행 방향), skipna = True 기본설정
print(df.mean(axis=0,
              skipna=False)) # NaN은 연산이 안돼요
```
