```python
import pandas as pd
import numpy as np

df = pd.read_csv('./data/auto-mpg.csv', header=None)

df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower',
              'weight', 'acceleration', 'model year', 'origin', 'name']

display(df)
```

# 기초

```python
# 1. 데이터프레임의 전체 행(row)과 열(column)의 개수를 확인하시오.
print(df.shape)
```

```python
# 2. origin 컬럼에 존재하는 서로 다른 값들은 무엇인지 출력하시오.
df['origin'].unique()
```

```python
# 3. cylinders 컬럼에 어떤 값들이 존재하며, 각각 몇 번 등장하는지 세시오.
df['cylinders'].value_counts()
```

```python
# 4. 데이터셋에서 가장 앞의 7행을 출력하시오.
df.head(7)
```

```python
# 5. model year 컬럼의 고유값이 몇 개인지 확인하시오.
len(df['model year'].unique())
```

# 응용

```python
# 1. 결측치(NaN)가 있는 컬럼이 몇 개인지, 각각 몇 개의 결측치가 있는지 확인하시오.
# df.isna().sum() >0
# df['horsepower'].replace('?',
#                          np.nan,
#                          inplace=True)
df['horsepower'] = df['horsepower'].replace('?', np.nan)
df.isna().sum()[df.isna().sum() > 0]
```

```python
# 2. origin이 1(미국) 또는 2(유럽)인 차량만 필터링하여 mpg, origin, name 컬럼을 출력하시오.
df.loc[df['origin'].isin([1,2]), ['mpg', 'origin', 'name']]
```

```python
# 3. weight 기준으로 차량을 오름차순 정렬한 후 상위 10개의 차량 이름과 무게를 출력하시오.

df[['name', 'weight']].sort_values(by='weight').head(10)
```

```python
# 4. horsepower가 100 이상이면서 cylinders가 4개인 차량의 개수를 세시오.
mask = (df['horsepower'].astype(np.float64) >= 100) & (df['cylinders'] == 4)
df[mask].shape[0]
```

```python
# 5. acceleration 컬럼의 평균을 구하시오. (결측치는 제외)
print(df['acceleration'].mean()) # 15.568090452261307
```
